```python
from sklearn.preprocessing import QuantileTransformer
sd = QuantileTransformer()


df['VLDL_st'] = sd.fit_transform(df[['VLDL']])
plt.figure(figsize=(15,8))
sns.histplot(data=df,x="VLDL_st",kde=True,hue="Gender", alpha = 0.6, bins = 12)
#plt.yscale('log')
plt.legend(title='Gender', labels=['Female', 'Male'])
plt.title('VLDL...')
sns.despine()
plt.grid()
plt.tight_layout()
plt.show()
```
